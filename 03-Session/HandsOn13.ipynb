## Introdução ao NumPy: O Pilar Numérico da Computação Científica

Bem-vindos à Sessão 3! Hoje vamos conhecer uma das bibliotecas mais importantes para a computação científica em Python: o **NumPy**.

NumPy (Numerical Python) é a base para quase todas as ferramentas de computação numérica em Python, incluindo Pandas, Matplotlib e scikit-learn. Ele fornece um objeto de array multidimensional de alto desempenho e ferramentas para trabalhar com esses arrays.

### Por que o NumPy é Essencial?

1.  **Arrays Eficientes:** O principal objeto do NumPy é o `ndarray` (N-dimensional array). Estes arrays são muito mais eficientes em termos de armazenamento e desempenho do que as listas Python para operações numéricas.
2.  **Operações Vetorizadas:** O NumPy permite realizar operações matemáticas em arrays inteiros de uma só vez, sem a necessidade de ciclos explícitos. Isto torna o código mais conciso, legível e, crucialmente, muito mais rápido.
3.  **Base para a Ciência de Dados:** Muitas bibliotecas científicas e de análise de dados em Python são construídas sobre o NumPy. Aprender NumPy é um passo fundamental para trabalhar com dados numéricos de forma eficaz.

Vamos ver um pequeno exemplo da diferença de velocidade entre listas Python e arrays NumPy.


```python
import numpy as np
import time

# Exemplo 1: Soma de elementos em listas Python vs. arrays NumPy
tamanho_lista = 10**6 # Um milhão de elementos

# Criar listas Python
lista_a = list(range(tamanho_lista))
lista_b = list(range(tamanho_lista))

# Criar arrays NumPy
array_a = np.arange(tamanho_lista) # np.arange é como range, mas para arrays NumPy
array_b = np.arange(tamanho_lista)

# Operação com listas Python
inicio_listas = time.time()
soma_listas = [lista_a[i] + lista_b[i] for i in range(tamanho_lista)]
fim_listas = time.time()
tempo_listas = fim_listas - inicio_listas
print(f"Tempo para somar listas Python: {tempo_listas:.6f} segundos")

# Operação com arrays NumPy (vetorizada)
inicio_numpy = time.time()
soma_numpy = array_a + array_b # Operação vetorizada!
fim_numpy = time.time()
tempo_numpy = fim_numpy - inicio_numpy
print(f"Tempo para somar arrays NumPy: {tempo_numpy:.6f} segundos")

print(f"\nNumPy é aproximadamente {tempo_listas / tempo_numpy:.1f} vezes mais rápido!")
```

Como podem ver, para operações numéricas em grandes conjuntos de dados, o NumPy é significativamente mais rápido. Esta eficiência é fundamental para a computação científica.

### O Objeto `ndarray`

O `ndarray` é o coração do NumPy. É um contentor para dados genéricos multidimensionais.

* **Homogéneo:** Todos os elementos de um `ndarray` devem ser do mesmo tipo de dados (ex: todos inteiros, todos floats). Isto é o que o torna eficiente.
* **Dimensões:** Pode ter 0 dimensões (escalar), 1 dimensão (vetor), 2 dimensões (matriz), ou mais (tensor).

Vamos criar o nosso primeiro array NumPy.

```python
# Criar um array a partir de uma lista Python
meu_primeiro_array = np.array([1, 2, 3, 4, 5])
print(f"Meu primeiro array: {meu_primeiro_array}")
print(f"Tipo do objeto: {type(meu_primeiro_array)}")
print(f"Tipo de dados dos elementos (dtype): {meu_primeiro_array.dtype}")
print(f"Número de dimensões (ndim): {meu_primeiro_array.ndim}")
print(f"Forma do array (shape): {meu_primeiro_array.shape}") # (5,) significa 5 elementos em 1 dimensão

# Criar um array com floats
array_floats = np.array([1.0, 2.5, 3.7])
print(f"\nArray de floats: {array_floats}")
print(f"Tipo de dados dos elementos (dtype): {array_floats.dtype}")
```
