### Exercício 4 (Matemática/Física): Simulação de Posição de Projétil

* `velocidade_inicial_y = 20` (m/s)

* $x(t) = velocidade\_inicial\_x * t$

**Tarefa:**
1.  Criem o array `tempo` usando `np.linspace`.
2.  Calculem o array de posições `x` para cada ponto no tempo.
3.  Calculem o array de posições `y` para cada ponto no tempo.
4.  Combinem os arrays `x` e `y` numa única matriz 2D onde a primeira coluna é `x` e a segunda coluna é `y`. A forma final deve ser `(11, 2)`.
5.  Imprimam a matriz de posições.
