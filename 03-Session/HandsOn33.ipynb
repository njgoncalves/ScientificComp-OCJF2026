## Operações Vetorizadas: Aritmética, Funções Matemáticas e Lógicas em Arrays

Uma das maiores vantagens do NumPy é a sua capacidade de realizar **operações vetorizadas**. Isto significa que podemos aplicar operações matemáticas, lógicas ou outras funções a arrays inteiros de uma só vez, sem a necessidade de escrever ciclos explícitos. Isto não só torna o código mais conciso e legível, mas também drasticamente mais rápido.

```python
import numpy as np
```

### 1. Aritmética Básica com Arrays

As operações aritméticas (`+`, `-`, `*`, `/`, `**`) são aplicadas elemento a elemento quando se trabalha com arrays NumPy.

#### 1.1. Operações entre um Array e um Escalar

Quando se opera um array com um único número (escalar), a operação é aplicada a cada elemento do array.

```python
vetor = np.array([1, 2, 3, 4, 5])

print(f"Vetor original: {vetor}")
print(f"Vetor + 10: {vetor + 10}") # Soma 10 a cada elemento
print(f"Vetor * 2: {vetor * 2}")   # Multiplica cada elemento por 2
print(f"Vetor / 5: {vetor / 5}")   # Divide cada elemento por 5
print(f"Vetor ao quadrado: {vetor ** 2}") # Eleva cada elemento ao quadrado
```

#### 1.2. Operações entre Dois Arrays

Quando se opera entre dois arrays, as operações são realizadas elemento a elemento, desde que os arrays tenham a mesma forma (ou que uma das formas possa ser "transmitida" - *broadcasting*).

**Exemplo:**

```python
vetor1 = np.array([1, 2, 3])
vetor2 = np.array([10, 20, 30])

print(f"Vetor 1: {vetor1}")
print(f"Vetor 2: {vetor2}")

print(f"Soma elemento a elemento: {vetor1 + vetor2}")
print(f"Multiplicação elemento a elemento: {vetor1 * vetor2}")

# Exemplo com matrizes
matriz1 = np.array([[1, 2], [3, 4]])
matriz2 = np.array([[5, 6], [7, 8]])

print(f"\nMatriz 1:\n{matriz1}")
print(f"Matriz 2:\n{matriz2}")

print(f"Soma de matrizes elemento a elemento:\n{matriz1 + matriz2}")
```

### 2. Funções Matemáticas Universais (ufuncs)

NumPy oferece uma vasta gama de funções matemáticas que operam elemento a elemento nos arrays. Estas são chamadas de "Universal Functions" (ufuncs).

**Exemplos:**
* `np.sqrt()`: Raiz quadrada
* `np.sin()`, `np.cos()`, `np.tan()`: Funções trigonométricas
* `np.log()`, `np.exp()`: Logaritmo natural e exponencial
* `np.abs()`: Valor absoluto
* `np.ceil()`, `np.floor()`, `np.round()`: Arredondamento

**Exemplos:**

```python
valores = np.array([4, 9, 16, 25])
print(f"Valores: {valores}")
print(f"Raiz quadrada: {np.sqrt(valores)}")

angulos_rad = np.array([0, np.pi/2, np.pi]) # np.pi é o valor de Pi
print(f"\nÂngulos em radianos: {angulos_rad}")
print(f"Seno dos ângulos: {np.sin(angulos_rad)}")

decimais = np.array([1.2, 2.8, 3.5, 4.1])
print(f"\nDecimais: {decimais}")
print(f"Arredondado para o inteiro mais próximo: {np.round(decimais)}")
print(f"Arredondado para cima (ceil): {np.ceil(decimais)}")
print(f"Arredondado para baixo (floor): {np.floor(decimais)}")
```

### 3. Funções de Agregação

Estas funções calculam um único valor a partir de um array.

* `np.sum()`: Soma de todos os elementos.
* `np.mean()`: Média dos elementos.
* `np.min()`, `np.max()`: Valor mínimo e máximo.
* `np.std()`: Desvio padrão.
* `np.var()`: Variância.

Podem ser aplicadas a todo o array ou ao longo de um eixo específico (`axis`).

**Exemplos:**

```python
medicoes = np.array([10, 12, 11, 13, 10, 14])

print(f"Medições: {medicoes}")
print(f"Soma total: {np.sum(medicoes)}")
print(f"Média: {np.mean(medicoes)}")
print(f"Valor mínimo: {np.min(medicoes)}")
print(f"Valor máximo: {np.max(medicoes)}")
print(f"Desvio padrão: {np.std(medicoes):.2f}") # Formata para 2 casas decimais

# Exemplo com matriz e eixo
matriz_notas = np.array([
    [15, 18, 12], # Notas do Aluno A
    [10, 14, 16]  # Notas do Aluno B
])
print(f"\nMatriz de notas:\n{matriz_notas}")

print(f"Média de todas as notas: {np.mean(matriz_notas):.2f}")
mean_axis = np.mean(matriz_notas, axis=1)
print(f"Média por aluno (ao longo das colunas, axis=1): {np.mean(matriz_notas, axis=1):.2f}")
print(f"Média por disciplina (ao longo das linhas, axis=0): {np.mean(matriz_notas, axis=0):.2f}")
```

### 4. Operadores Lógicos e Comparação

Operações de comparação (`>`, `<`, `==`, `!=`, etc.) e lógicas (`&` para AND, `|` para OR, `~` para NOT) também são vetorizadas e retornam arrays booleanos.

**Exemplos:**

```python
# Célula de Código
alturas = np.array([1.65, 1.78, 1.70, 1.85, 1.60])
print(f"Alturas: {alturas}")

print(f"Alturas maiores que 1.70m: {alturas > 1.70}")
print(f"Alturas iguais a 1.70m: {alturas == 1.70}")

# Combinando condições
# Alturas entre 1.65m e 1.75m (inclusive)
condicao_complexa = (alturas >= 1.65) & (alturas <= 1.75)
print(f"Alturas entre 1.65m e 1.75m: {condicao_complexa}")

# Usando o array booleano para filtrar
alturas_filtradas = alturas[condicao_complexa]
print(f"Alturas filtradas: {alturas_filtradas}")
```
