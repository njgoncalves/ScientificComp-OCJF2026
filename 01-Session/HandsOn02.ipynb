### 1. Variáveis: Caixas para Guardar Informação

Pensem nas **variáveis** como caixas onde podemos guardar diferentes tipos de informação. Damos um nome a cada caixa para podermos aceder ao que lá está guardado mais tarde.

Para criar uma variável em Python, basta dar-lhe um nome e atribuir um valor usando o sinal de igual (`=`).

**Exemplo:**

```python
nome_aluno = "Ana Silva"
idade = 17
nota_fisica = 15.5
disciplina_preferida = "Biologia"

# Podemos "imprimir" o valor de uma variável para o ver:
print(nome_aluno)
print(idade)
print(nota_fisica)
print(disciplina_preferida)
```

**Regras para nomes de variáveis:**
* Devem começar com uma letra (a-z, A-Z) ou um `_` (underscore).
* Não podem começar com um número.
* Podem conter letras, números e `_`.
* São *case-sensitive* (maiúsculas e minúsculas importam): `nota` é diferente de `Nota`.
* Evitem usar palavras reservadas do Python (ex: `print`, `if`, `for`).

### 2. Tipos de Dados: Que Tipo de Informação Guardamos?

Cada informação que guardamos numa variável tem um **tipo**. Python é uma linguagem de "tipagem dinâmica", o que significa que não precisamos de dizer qual é o tipo da variável, o Python descobre automaticamente.

Os tipos de dados mais comuns são:

* **Inteiros (`int`):** Números inteiros (positivos, negativos ou zero). Ex: `10`, `-5`, `0`.
* **Ponto flutuante (`float`):** Números com casas decimais. Ex: `3.14`, `-0.5`, `10.0`.
* **Strings (`str`):** Sequências de caracteres (texto). Devem ser delimitadas por aspas simples (`'`) ou duplas (`"`). Ex: `"Olá mundo"`, `'Python é fixe'`.
* **Booleanos (`bool`):** Representam verdadeiro ou falso. Os únicos valores são `True` ou `False` (com a primeira letra maiúscula!).

Podemos usar a função `type()` para verificar o tipo de uma variável:

```python
numero_inteiro = 25
numero_decimal = 3.14159
frase = "A ciência é divertida!"
verdadeiro_ou_falso = True

print(type(numero_inteiro))
print(type(numero_decimal))
print(type(frase))
print(type(verdadeiro_ou_falso))
```

### 3. Operadores: Fazer Coisas com os Dados

Os operadores permitem-nos realizar operações com os valores das nossas variáveis.

#### 3.1. Operadores Aritméticos

Usados para cálculos matemáticos:

* `+`: Adição
* `-`: Subtração
* `*`: Multiplicação
* `/`: Divisão (sempre retorna um `float`)
* `//`: Divisão inteira (descarta a parte decimal)
* `%`: Módulo (resto da divisão)
* `**`: Exponenciação (potência)

**Exemplos:**

```python
a = 10
b = 3

print(f"Adição: {a + b}")
print(f"Subtração: {a - b}")
print(f"Multiplicação: {a * b}")
print(f"Divisão: {a / b}")
print(f"Divisão Inteira: {a // b}")
print(f"Módulo (Resto): {a % b}")
print(f"Exponenciação (a elevado a b): {a ** b}")

# f-string: uma forma fácil de incluir variáveis dentro de strings. O "f" antes das aspas indica que é uma f-string.
```

#### 3.2. Operadores de Atribuição

Usados para atribuir valores a variáveis. O `+=`, `-=`, etc. são atalhos para operações comuns:

* `=` : Atribuição simples
* `+=` : `x = x + y` é o mesmo que `x += y`
* `-=` : `x = x - y` é o mesmo que `x -= y`
* `*=` : `x = x * y` é o mesmo que `x *= y`
* `/=` : `x = x / y` é o mesmo que `x /= y`

**Exemplos:**

```python
x = 5
print(f"Valor inicial de x: {x}")

x += 3 # É o mesmo que x = x + 3
print(f"x depois de x += 3: {x}")

x *= 2 # É o mesmo que x = x * 2
print(f"x depois de x *= 2: {x}")
```

#### 3.3. Operadores de Comparação

Usados para comparar valores. Retornam sempre um valor booleano (`True` ou `False`).

* `==`: Igual a
* `!=`: Diferente de
* `>` : Maior que
* `<` : Menor que
* `>=`: Maior ou igual a
* `<=`: Menor ou igual a

**Exemplos:**

```python
idade_joao = 20
idade_maria = 22

print(f"João tem a mesma idade que Maria? {idade_joao == idade_maria}")
print(f"João é mais velho que Maria? {idade_joao > idade_maria}")
print(f"Maria é mais velha ou igual que João? {idade_maria >= idade_joao}")
```

#### 3.4. Operadores Lógicos

Usados para combinar condições booleanas:

* `and`: Verdadeiro se *ambas* as condições forem verdadeiras.
* `or`: Verdadeiro se *pelo menos uma* das condições for verdadeira.
* `not`: Inverte o valor booleano (verdadeiro torna-se falso e vice-versa).

**Exemplos:**

```python
tem_passagem = True
chegou_a_horas = False
esta_chover = True

print(f"Pode entrar (tem passagem E chegou a horas)? {tem_passagem and chegou_a_horas}")
print(f"Pode sair (tem passagem OU não está a chover)? {tem_passagem or not esta_chover}")
print(f"Não está a chover? {not esta_chover}")
```

### 4. Conversão de Tipos (Type Casting)

Às vezes, precisamos de converter um tipo de dado para outro. Por exemplo, converter um número que está como `string` para um `int` para poder fazer cálculos.

* `int()`: Converte para inteiro.
* `float()`: Converte para float.
* `str()`: Converte para string.

**Exemplos:**

```python
numero_str = "123"
numero_float = 45.67

# Converter string para int
numero_int = int(numero_str)
print(f"'{numero_str}' convertido para int: {numero_int}, Tipo: {type(numero_int)}")

# Converter float para int (perde as casas decimais!)
inteiro_do_float = int(numero_float)
print(f"{numero_float} convertido para int: {inteiro_do_float}, Tipo: {type(inteiro_do_float)}")

# Converter int para float
flutuante_do_int = float(numero_int)
print(f"{numero_int} convertido para float: {flutuante_do_int}, Tipo: {type(flutuante_do_int)}")

# Converter número para string
texto_do_numero = str(numero_float)
print(f"{numero_float} convertido para string: '{texto_do_numero}', Tipo: {type(texto_do_numero)}")
```

**Cuidado:** Não tentem converter texto que não seja um número para `int` ou `float`! Irá dar um erro.
