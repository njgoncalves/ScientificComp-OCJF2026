### Exercício 1: Cálculo de IMC (Índice de Massa Corporal)

O Índice de Massa Corporal (IMC) é calculado pela fórmula: $IMC = peso / (altura^2)$.

**Tarefa:**
1.  Criem uma variável `peso_kg` com um valor numérico (ex: 70.5).
2.  Criem uma variável `altura_metros` com um valor numérico (ex: 1.75).
3.  Calculem o IMC e guardem o resultado numa variável `imc`.
4.  Imprimam o valor do IMC.

### Exercício 2: Manipulação de Strings e Concatenção

**Tarefa:**
1.  Criem uma variável `nome` com o vosso primeiro nome.
2.  Criem uma variável `apelido` com o vosso apelido.
3.  Criem uma variável `nome_completo` que seja a concatenação do `nome` e `apelido`, separados por um espaço.
4.  Imprimam o `nome_completo`.
5.  Imprimam uma mensagem a dizer "O meu nome completo tem X caracteres.", onde X é o número de caracteres do `nome_completo` (usem a função `len()`).

### Exercício 3: Operadores de Comparação e Lógicos

**Tarefa:**
Temos dois valores que representam a temperatura de uma amostra em diferentes momentos.

* `temperatura_inicial = 25.0`
* `temperatura_final = 28.5`

1.  Verifiquem se a `temperatura_final` é maior que a `temperatura_inicial`. Imprimam o resultado.
2.  Verifiquem se a `temperatura_final` é igual a `30.0`. Imprimam o resultado.
3.  Verifiquem se a `temperatura_final` é maior que `temperatura_inicial` **E** se a `temperatura_final` é menor que `30.0`. Imprimam o resultado.

### Exercício 4: Conversão de Tipos e Cálculos com Datas

Imaginem que recebem o ano de nascimento de um utilizador como uma string.

**Tarefa:**
1.  Criem uma variável `ano_nascimento_str = "2007"`.
2.  Criem uma variável `ano_atual` com o ano corrente (2025).
3.  Convertam `ano_nascimento_str` para um tipo numérico apropriado.
4.  Calculem a idade do utilizador e guardem-na numa variável `idade`.
5.  Imprimam a idade.

### Exercício 5: Média de Notas (Com Entrada Simulada)

Imaginem que querem calcular a média de 3 notas de um aluno.

**Tarefa:**
1.  Criem variáveis `nota1 = 14`, `nota2 = 17`, `nota3 = 12`.
2.  Calculem a `media_notas`.
3.  Imprimam a média.
4.  Arredondem a média para duas casas decimais (podem usar a função `round()`).
5.  Imprimam a média arredondada.
