## Boas-vindas ao Jupyter Notebook!

Neste *hands-on*, vamos explorar a ferramenta que será o nosso laboratório de programação ao longo deste curso: o **Jupyter Notebook**.

O Jupyter Notebook é um ambiente interativo onde podemos combinar código Python, texto, equações e visualizações numa única página web. É perfeito para a computação científica, pois permite-nos documentar o nosso trabalho passo a passo, experimentar ideias e partilhar os nossos resultados de forma clara.

Vamos começar!

### 1. Tipos de Células: Código e Markdown

O Jupyter Notebook é composto por **células**. Existem dois tipos principais de células que usaremos:

* **Células de Código:** Contêm código Python que pode ser executado.
* **Células de Markdown:** Contêm texto formatado usando a linguagem Markdown, que permite títulos, listas, negrito, itálico, etc.

**Experimentem:**

1.  Cliquem na célula abaixo. Ela é uma célula de **código**.
2.  Para executar o código, podem clicar no botão "Run" (um triângulo) na barra de ferramentas ou pressionar `Shift + Enter`.

```python
print("Olá, Jupyter!")
print("Esta é a minha primeira linha de código Python.")
```

**Agora, experimentem criar uma nova célula:**

1.  Cliquem no botão `+` na barra de ferramentas para adicionar uma nova célula.
2.  Por padrão, ela será uma célula de código.
3.  Podem mudar o tipo da célula para **Markdown** usando o menu *dropdown* na barra de ferramentas (onde diz "Code").
4.  Cliquem duas vezes nesta célula de Markdown para verem o código Markdown por trás dela e depois executem-na (`Shift + Enter`) para renderizar o texto.

### 2. Guardar e Abrir Notebooks

* Para **guardar** o vosso trabalho, cliquem no ícone do disquete (Save) na barra de ferramentas ou vão a `File > Save and Checkpoint`.
* Os Notebooks são guardados com a extensão `.ipynb`.
* Para **abrir** um Notebook existente, usem a interface do Jupyter (o vosso navegador web) para navegar até ao ficheiro `.ipynb` e cliquem nele.

### 3. Modos de Operação: Edição e Comando

O Jupyter Notebook tem dois modos principais de operação:

* **Modo de Edição (Edit Mode):** Quando estão a editar o conteúdo de uma célula (o cursor está a piscar dentro da célula). A borda da célula fica **verde**.
* **Modo de Comando (Command Mode):** Quando uma célula está selecionada, mas não estão a editá-la. A borda da célula fica **azul**. Neste modo, podem usar atalhos de teclado para manipular células inteiras.

**Experimentem:**

1.  Cliquem numa célula para a selecionar (modo de comando - borda azul).
2.  Pressionem `Enter` para entrar no modo de edição (borda verde).
3.  Pressionem `Esc` para voltar ao modo de comando.

### 4. Atalhos de Teclado Úteis (Modo de Comando)

Aprender alguns atalhos pode acelerar muito o vosso trabalho:

* **`A`**: Adicionar uma célula **A**cima da célula selecionada.
* **`B`**: Adicionar uma célula a**B**aixo da célula selecionada.
* **`DD`** (pressionar `D` duas vezes): **D**eletar a célula selecionada.
* **`M`**: Mudar o tipo da célula selecionada para **M**arkdown.
* **`Y`**: Mudar o tipo da célula selecionada para CÓDIGO (p**Y**thon).
* **`Z`**: Desfazer a última ação (por exemplo, apagar uma célula).
* **`Shift + Enter`**: Executar a célula e ir para a próxima (ou criar uma nova se for a última).
* **`Ctrl + Enter`**: Executar a célula e permanecer na mesma.

**Experimentem:**

* Usem `A` e `B` para adicionar células.
* Usem `DD` para apagar as células que acabaram de criar.
* Mudem o tipo de uma célula para Markdown e depois para Código usando `M` e `Y`.

### 5. Reiniciar o Kernel (Quando as Coisas Correm Mal)

O "kernel" é o motor que executa o código Python. Por vezes, o kernel pode ficar "preso" ou precisar de ser reiniciado para limpar as variáveis da memória.

* Vão a `Kernel > Restart` (ou `Kernel > Restart & Clear Output` para limpar as saídas de todas as células).
* Isto é útil se o vosso código estiver a dar problemas estranhos ou se quiserem começar "do zero" com as variáveis.
