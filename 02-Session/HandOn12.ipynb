## Estruturas Condicionais: Tomada de Decisões no Código

Bem-vindos à segunda sessão! Hoje vamos aprender a dar ao nosso código a capacidade de "tomar decisões". Isto é feito através das **estruturas condicionais**, que permitem que diferentes blocos de código sejam executados dependendo se uma condição é verdadeira ou falsa.

Vamos explorar o `if`, `elif` e `else`.

### 1. A Declaração `if`: "Se isto for verdade..."

A forma mais básica de uma estrutura condicional é o `if`. O código dentro do `if` só será executado se a condição for `True`.

**Sintaxe:**
```python
if condicao_aqui:
    # Código que será executado se a condição for True
    # Nota a indentação! É crucial em Python.
```

* A `condicao_aqui` deve ser uma expressão que resulta num valor booleano (`True` ou `False`).
* Os dois pontos (`:`) são obrigatórios no final da linha `if`.
* O código dentro do `if` deve estar **indentado** (geralmente com 4 espaços ou uma tabulação). A indentação define o bloco de código.

**Exemplo:** Verificar se um aluno passou de ano.

````python
nota_final = 14

if nota_final >= 10:
    print("Parabéns! O aluno foi aprovado.")
    print("Pode seguir para o próximo ano letivo.")

print("Fim da verificação de aprovação.") # Esta linha executa sempre, pois não está indentada no 'if'
````

### 2. A Declaração `else`: "...Senão, faz isto."

O `else` permite-nos especificar um bloco de código a ser executado quando a condição do `if` (ou do `elif` anterior) é `False`.

**Sintaxe:**
```python
if condicao:
    # Código se a condição for True
else:
    # Código se a condição for False
```

**Exemplo:** Verificar se um número é par ou ímpar.

````python
numero = 7

if numero % 2 == 0: # O operador % (módulo) dá o resto da divisão. Se o resto da divisão por 2 for 0, é par.
    print(f"O número {numero} é par.")
else:
    print(f"O número {numero} é ímpar.")
````

### 3. A Declaração `elif`: "Senão, se isto for verdade..."

Quando precisamos de verificar múltiplas condições em sequência, usamos o `elif` (abreviação de "else if"). O Python verifica as condições por ordem, e executa o primeiro bloco cujo `if` ou `elif` seja verdadeiro. Se nenhum for verdadeiro, e houver um `else`, este será executado.

**Sintaxe:**
```python
if condicao1:
    # Código se condicao1 for True
elif condicao2:
    # Código se condicao1 for False E condicao2 for True
elif condicao3:
    # Código se condicao1 e condicao2 forem False E condicao3 for True
else:
    # Código se nenhuma das condições anteriores for True
```

**Exemplo:** Classificar um número em positivo, negativo ou zero.

````python
valor = 0

if valor > 0:
    print(f"O valor {valor} é positivo.")
elif valor < 0:
    print(f"O valor {valor} é negativo.")
else:
    print(f"O valor {valor} é zero.")
````

### 4. Condições Múltiplas com Operadores Lógicos (`and`, `or`, `not`)

Podemos combinar múltiplas condições usando os operadores lógicos que aprendemos na Sessão 1: `and`, `or`, `not`.

**Exemplo:** Verificar se um aluno foi aprovado e teve uma nota alta.

````python
media = 15
presenca = 85 # percentagem

if media >= 10 and presenca >= 75:
    print("O aluno foi aprovado por média e presença.")
elif media >= 10 and presenca < 75:
    print("O aluno foi aprovado por média, mas reprovou por faltas.")
else:
    print("O aluno não foi aprovado.")
````
