## Funções: Definição, Argumentos e Retorno

À medida que os nossos programas crescem, começamos a ter blocos de código que realizam tarefas específicas e que podem ser repetidos. Para evitar reescrever o mesmo código várias vezes e para tornar os nossos programas mais organizados e fáceis de manter, usamos **funções**.

Uma função é um bloco de código reutilizável que realiza uma tarefa específica.

Vamos aprender a definir e usar funções em Python.

### 1. Definir uma Função: A Palavra-Chave `def`

Para definir uma função em Python, usamos a palavra-chave `def`, seguida do nome da função, parênteses `()` e dois pontos `:`. O código da função deve ser indentado.

**Sintaxe:**
```python
def nome_da_funcao():
    # Código que a função executa
    # Este código só será executado quando a função for "chamada"
```

**Exemplo:** Uma função simples que imprime uma saudação.

````python
def saudar():
    """Esta função imprime uma saudação simples."""
    print("Olá! Bem-vindo ao mundo das funções.")

# Para executar a função, temos de a "chamar" pelo nome, seguido de parênteses:
saudar()
saudar() # Podemos chamar a função quantas vezes quisermos
````

**Notas:**
* A string entre aspas triplas (`"""Docstring"""`) é uma **docstring**. É uma boa prática usá-las para descrever o que a função faz. É útil para documentação.
* A função `saudar()` não recebe nenhuma informação externa nem devolve nada.

### 2. Argumentos: Passar Informação para a Função

As funções tornam-se muito mais úteis quando podem receber informação para trabalhar. Essa informação é passada através de **argumentos** (também chamados de parâmetros). Os argumentos são definidos dentro dos parênteses na definição da função.

**Sintaxe:**
```python
def nome_da_funcao(argumento1, argumento2):
    # Código que usa argumento1 e argumento2
```

**Exemplo:** Uma função que saúda uma pessoa pelo nome.

````python
def saudar_nome(nome):
    """Esta função imprime uma saudação personalizada com o nome."""
    print(f"Olá, {nome}! É um prazer ter-te aqui.")

saudar_nome("Ana")
saudar_nome("Pedro")
````

Podemos ter múltiplos argumentos.

**Exemplo:** Calcular a área de um retângulo.

````python
def calcular_area_retangulo(largura, altura):
    """Calcula e imprime a área de um retângulo."""
    area = largura * altura
    print(f"A área do retângulo é: {area}")

calcular_area_retangulo(5, 10)
calcular_area_retangulo(3.5, 7)
````

### 3. Retorno de Valores: Devolver Resultados da Função

Muitas vezes, queremos que uma função não apenas faça algo, mas também nos **devolva um resultado** para que possamos usá-lo noutra parte do nosso programa. Para isso, usamos a palavra-chave `return`.

Quando uma função encontra um `return`, ela termina a sua execução e envia o valor especificado de volta para onde foi chamada.

**Sintaxe:**
```python
def nome_da_funcao(argumentos):
    # Código da função
    resultado = ... # Calcula o resultado
    return resultado # Devolve o resultado
```

**Exemplo:** A função de área do retângulo, mas agora devolvendo o valor.

````python
def calcular_area_retangulo_com_retorno(largura, altura):
    """Calcula a área de um retângulo e retorna o valor."""
    area = largura * altura
    return area

# Guardamos o resultado da função numa variável:
area1 = calcular_area_retangulo_com_retorno(5, 10)
print(f"A área do primeiro retângulo é: {area1}")

area2 = calcular_area_retangulo_com_retorno(3.5, 7)
print(f"A área do segundo retângulo é: {area2}")

total_areas = area1 + area2
print(f"A soma das duas áreas é: {total_areas}")
````

**Notas:**
* Uma função pode retornar qualquer tipo de dado (número, string, booleano, lista, tuplo, etc.).
* Se uma função não tiver uma declaração `return`, ela implicitamente retorna `None`.

### 4. Argumentos Padrão (Default Arguments)

Podemos dar valores padrão a argumentos. Se o utilizador não fornecer um valor para esse argumento quando chamar a função, o valor padrão será usado.

**Exemplo:**

````python
# Célula de Código
def saudar_com_idioma(nome, idioma="Português"):
    """Saúda uma pessoa num idioma específico (Português por padrão)."""
    if idioma == "Português":
        print(f"Olá, {nome}!")
    elif idioma == "Inglês":
        print(f"Hello, {nome}!")
    else:
        print(f"Desculpe, não conheço o idioma {idioma}.")

saudar_com_idioma("Maria") # Usa o idioma padrão (Português)
saudar_com_idioma("John", "Inglês")
saudar_com_idioma("Pierre", "Francês")
````
