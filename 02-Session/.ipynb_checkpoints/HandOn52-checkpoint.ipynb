## Exercícios Práticos: Aplicação Integrada dos Conceitos

Chegou a hora de integrar tudo o que aprendemos: estruturas condicionais, ciclos, listas, tuplos e funções! Estes exercícios foram desenhados para simular problemas que podem encontrar nas vossas disciplinas de ciências.

Lembrem-se de usar as células de código para as vossas soluções e de as executar para testar.

Boa sorte!

### Exercício 1 (Matemática): Classificação de Ângulos

**Tarefa:**
Escrevam uma função chamada `classificar_angulo` que receba um número (representando um ângulo em graus) como argumento. A função deve imprimir a classificação do ângulo:

* Se o ângulo for `90`: "Ângulo Reto"
* Se o ângulo for `> 0` e `< 90`: "Ângulo Agudo"
* Se o ângulo for `> 90` e `< 180`: "Ângulo Obtuso"
* Se o ângulo for `180`: "Ângulo Raso"
* Se o ângulo for `360`: "Ângulo Completo"
* Para qualquer outro valor: "Ângulo Inválido ou Outro Tipo"

**Dica:** Usem `if`, `elif` e `else`.

### Exercício 2 (Físico-Química): Cálculo de Concentração Molar

A concentração molar ($C$) de uma solução é dada pela fórmula: $C = n / V$, onde $n$ é o número de moles do soluto e $V$ é o volume da solução em litros.

**Tarefa:**
1.  Criem uma função chamada `calcular_concentracao_molar` que receba `moles` e `volume_litros` como argumentos.
2.  A função deve retornar a concentração molar.
3.  **Adicionem uma condição:** Se o `volume_litros` for `0`, a função deve imprimir uma mensagem de erro (ex: "Erro: O volume não pode ser zero!") e retornar `None`.
4.  Testem a função com valores válidos e com `volume_litros = 0`.

### Exercício 3 (Biologia e Geologia): Simulação de Crescimento Populacional Simples

Vamos simular um crescimento populacional simples ao longo de alguns anos.

**Tarefa:**
1.  Criem uma lista chamada `populacao_inicial` com o número de indivíduos em cada ano (ex: `[100, 110, 121, 133, 146]`).
2.  Usem um ciclo `for` para percorrer a lista e imprimir o número de indivíduos em cada ano, juntamente com o número do ano (começando no Ano 1).
    * Exemplo de saída: "Ano 1: 100 indivíduos"
3.  **Desafio:** Calculem a taxa de crescimento anual (percentagem) entre cada ano consecutivo e imprimam-na.
    * Fórmula: `((populacao_ano_seguinte - populacao_ano_atual) / populacao_ano_atual) * 100`
    * **Dica:** Vão precisar de aceder a dois elementos da lista por vez. O ciclo deve ir até `len(lista) - 1`.

### Exercício 4 (Matemática): Verificação de Número Primo

Um número primo é um número natural maior que 1 que não tem divisores positivos além de 1 e de si próprio.

**Tarefa:**
1.  Escrevam uma função chamada `e_primo` que receba um número inteiro como argumento.
2.  A função deve retornar `True` se o número for primo e `False` caso contrário.
    * **Dicas:**
        * Números menores ou iguais a 1 não são primos.
        * Podem usar um ciclo `for` para verificar divisores de `2` até `numero - 1`.
        * Se encontrarem qualquer divisor, o número não é primo (`return False`).
        * Se o ciclo terminar sem encontrar divisores, o número é primo (`return True`).
3.  Testem a função com vários números (ex: 7, 10, 2, 1).

### Exercício 5 (Físico-Química): Análise de Dados de pH

Temos uma lista de medições de pH de várias amostras de água.

**Tarefa:**
1.  Criem uma lista chamada `medicoes_ph` com alguns valores (ex: `[7.2, 6.8, 8.1, 5.5, 7.0, 7.5]`).
2.  Usem um ciclo `for` para percorrer a lista. Para cada valor de pH:
    * Se `pH < 7.0`: Imprimam "Amostra Ácida: pH X"
    * Se `pH > 7.0`: Imprimam "Amostra Básica: pH X"
    * Se `pH == 7.0`: Imprimam "Amostra Neutra: pH X"
3.  **Desafio:** Criem uma função `analisar_ph(lista_ph)` que faça esta análise e imprima os resultados.

### Exercício 6 (Biologia e Geologia): Contagem de Elementos em Tuplo de Minerais

Imaginem que têm um tuplo que representa uma sequência de minerais encontrados numa amostra geológica.

**Tarefa:**
1.  Criem um tuplo chamado `amostra_minerais` (ex: `("Quartzo", "Feldspato", "Mica", "Quartzo", "Calcite", "Quartzo")`).
2.  Usem um ciclo `for` para percorrer o tuplo e contem quantas vezes o mineral "Quartzo" aparece.
3.  Imprimam o resultado final (ex: "O Quartzo aparece X vezes na amostra.").
4.  **Desafio:** Criem uma função `contar_mineral(tuplo_minerais, mineral_alvo)` que receba o tuplo e o mineral a contar, e retorne a contagem.
