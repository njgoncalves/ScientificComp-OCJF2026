## Estruturas de Repetição: Automação de Tarefas Repetitivas

Nesta secção, vamos aprender a fazer o nosso código repetir ações. As **estruturas de repetição** (ou ciclos/loops) são incrivelmente úteis para automatizar tarefas que precisam de ser realizadas várias vezes.

Vamos explorar os ciclos `for` e `while`.

### 1. O Ciclo `for`: "Para cada item nesta sequência..."

O ciclo `for` é usado para iterar (percorrer) uma sequência de elementos, como uma lista de números, os caracteres de uma string, ou um intervalo de números.

**Sintaxe:**
```python
for item in sequencia:
    # Código a ser executado para cada item na sequência
    # A indentação é, novamente, crucial.
```

#### 1.1. Iterar sobre uma String

Uma string é uma sequência de caracteres.

````python
palavra = "Python"

print("A percorrer a palavra:")
for letra in palavra:
    print(letra)
````

#### 1.2. Iterar sobre um Intervalo de Números com `range()`

A função `range()` é muito útil para gerar sequências de números.

* `range(fim)`: Gera números de `0` até `fim-1`.
* `range(inicio, fim)`: Gera números de `inicio` até `fim-1`.
* `range(inicio, fim, passo)`: Gera números de `inicio` até `fim-1`, saltando de `passo` em `passo`.

**Exemplos:**

````python
print("Números de 0 a 4:")
for i in range(5): # Gera 0, 1, 2, 3, 4
    print(i)

print("\nNúmeros de 2 a 5:")
for num in range(2, 6): # Gera 2, 3, 4, 5
    print(num)

print("\nNúmeros pares de 0 a 10:")
for j in range(0, 11, 2): # Gera 0, 2, 4, 6, 8, 10
    print(j)
````

### 2. O Ciclo `while`: "Enquanto esta condição for verdadeira..."

O ciclo `while` executa um bloco de código repetidamente **enquanto uma condição for verdadeira**. É crucial que a condição se torne falsa em algum momento, caso contrário, teremos um "loop infinito".

**Sintaxe:**
```python
while condicao_aqui:
    # Código a ser executado enquanto a condição for True
    # É preciso ter algo aqui que mude a condição para False,
    # senão o ciclo nunca para!
```

**Exemplo:** Contar de 1 a 5.

````python
contador = 1
print("A contar de 1 a 5 com while:")
while contador <= 5:
    print(contador)
    contador += 1 # Incrementa o contador para que a condição eventualmente se torne falsa
````

**Experimentem:**

### 3. `break` e `continue`: Controlando os Ciclos

* **`break`:** Interrompe imediatamente o ciclo atual e o programa continua a execução na primeira linha após o ciclo.
* **`continue`:** Salta o resto do código na iteração atual e passa para a próxima iteração do ciclo.

**Exemplo com `break`:** Encontrar o primeiro número par.

````python
print("Encontrar o primeiro número par entre 1 e 10:")
for num in range(1, 11):
    if num % 2 == 0:
        print(f"O primeiro número par é: {num}")
        break # Sai do ciclo assim que encontra o primeiro par
    print(f"Verificando {num}...") # Esta linha só é impressa se 'break' não for acionado
````
