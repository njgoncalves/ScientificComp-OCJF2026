## Estudo de Caso Simples: Análise de Dados de Sensor Ambiental

Nesta sessão final, vamos juntar tudo o que aprendemos! O objetivo é resolver um problema de computação científica que requer a utilização integrada do **Pandas** para manipulação de dados, do **NumPy** para operações numéricas e do **Matplotlib** para visualização.

**Cenário:** Um sensor ambiental registou a temperatura e a humidade relativa ao longo de um dia. Queremos analisar estes dados para identificar tendências, calcular médias e visualizar as variações.

Vamos começar por importar as bibliotecas necessárias.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io # Para simular a leitura de CSV
```

### 1. Carregar e Inspecionar os Dados (Pandas)

Os dados do sensor foram guardados num ficheiro CSV. Vamos simulá-lo aqui.

```python
csv_dados_ambientais = """Hora,Temperatura_C,Humidade_Relativa_perc
0,18.5,80
2,17.9,85
4,17.2,88
6,17.0,90
8,18.8,75
10,21.5,65
12,24.0,55
14,26.5,48
16,25.8,50
18,23.1,60
20,20.5,70
22,19.0,78
"""

# Carregar os dados para um DataFrame
df_ambiental = pd.read_csv(io.StringIO(csv_dados_ambientais))

print("DataFrame de Dados Ambientais:\n", df_ambiental)
print("\nInformações do DataFrame:")
df_ambiental.info()
print("\nEstatísticas Descritivas:")
print(df_ambiental.describe())
```
